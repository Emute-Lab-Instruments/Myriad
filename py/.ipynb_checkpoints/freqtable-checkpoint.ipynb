```python
octaves=10
m = arange(4096) / 4096.0 * octaves
```

```python
plot(m)
```

```python
freqs = [pow(2,x) for x in m]
plot(freqs)
```

```python
print(freqs)
```

```python
cpufreq = 125000000
clockdiv=8
clockfreq = cpufreq/clockdiv
print("clockfreq", clockfreq)
wavelen20hz = clockfreq / 20
print("20hz", wavelen20hz)
```
